```python
from Bio import SeqIO
aligned_list = []
names = []
for seq_record in SeqIO.parse("out58_nospaces_beta.fasta", "fasta"):
    aligned_list.append(seq_record.seq)
    names.append(" ".join(seq_record.name.split('_')))
names = np.asarray(names)
names[names=='Clostridioides difficile'] = 'Clostridium difficile'
names[names=='Corynebacterium sp.'] = 'Corynebacterium'
names[names=='Mycolicibacterium fortuitum'] = 'Mycobacterium fortuitum'
names[names=='Cutibacterium acnes'] = 'Propionibacterium acnes'
```

```python
from dtaidistance import dtw
from Bio import pairwise2

df_58_norm_shifted = pd.read_csv('df_58_norm_shifted.csv', index_col=0)

table1 = pd.DataFrame()
table1['Pair'] = range(1,16)
table1['Organism 1'] = [i[0] for i in euclid_duos[:15]]
table1['Organism 2'] = [i[1] for i in euclid_duos[:15]]
table1['ED'] = np.round(pd.read_csv('distances_euc.csv').iloc[:,1].values[:15],2)
table1['ID'] = [round(100*(1-(max([len(aligned_list[np.where(names==duo[0])[0][0]]),len(aligned_list[np.where(names==duo[1])[0][0]])])-
 pairwise2.align.localxx(aligned_list[np.where(names==duo[0])[0][0]], aligned_list[np.where(names==duo[1])[0][0]], score_only=True))/max([len(aligned_list[np.where(names==duo[0])[0][0]]),len(aligned_list[np.where(names==duo[1])[0][0]])])),2) for duo in euclid_duos[:15]]
table1['Mismatches'] = [int(max([len(aligned_list[np.where(names==duo[0])[0][0]]),len(aligned_list[np.where(names==duo[1])[0][0]])])-
 pairwise2.align.localxx(aligned_list[np.where(names==duo[0])[0][0]], aligned_list[np.where(names==duo[1])[0][0]], score_only=True)) for duo in euclid_duos[:15]]
# table1['DTW w=1'] = [dtw.distance(df_58_norm_shifted.loc[duo[0]].values, df_58_norm_shifted.loc[duo[1]].values, window=1) for duo in euclid_duos[:15]]
# table1['DTW w=2'] = [dtw.distance(df_58_norm_shifted.loc[duo[0]].values, df_58_norm_shifted.loc[duo[1]].values, window=2) for duo in euclid_duos[:15]]
# table1['DTW w=3'] = [dtw.distance(df_58_norm_shifted.loc[duo[0]].values, df_58_norm_shifted.loc[duo[1]].values, window=3) for duo in euclid_duos[:15]]
table1['DTW'] = [np.round(dtw.distance(df_58_norm_shifted.loc[duo[0]].values, df_58_norm_shifted.loc[duo[1]].values),2) for duo in euclid_duos[:15]]
table1['Sorted by DTW'] = table1['DTW'].sort_values().index.values+1
table1
```
